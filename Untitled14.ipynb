### 1. convert object variables: x41 and x45 to float 64

```python
import numpy as np
#pd.to_numeric(train['x41'].str.replace(r'[,.$]',''))
train['x41']=train['x41'].str.replace('$','').astype(np.float64)
test['x41']=test['x41'].str.replace('$','').astype(np.float64)
train['x45']=train['x45'].str.replace('%','').astype(np.float64)
test['x45']=test['x45'].str.replace('%','').astype(np.float64)
```

## First try in modeling - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

import sklearn
from sklearn.preprocessing import RobustScaler


y=train.y
cols = list(train)
cols.remove('y')
x =train[cols]



# from sklearn.preprocessing import Imputer
# imp = Imputer(missing_values='NaN', strategy='mean', axis=0)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25)


##x_test=RobustScaler(quantile_range=(25, 75)).fit_transform(x_test)
#x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
#x_train, x_test, y_train, y_test =cross_validate.train_test_split(x,y,test_size=0.3)
#print(len(ov_trainy),sum(ov_trainy))
                                                                            
classifier = RandomForestClassifier(max_depth=100, random_state=1,class_weight="balanced")  

classifier.fit(x_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(x_test)  
c_m = confusion_matrix(y_test,y_pred )
print('Logistic Regression:\nconfusion matrix\n', c_m,'\n\n')
ax=plt.matshow(c_m,cmap=plt.cm.Blues)
print('Confusion matrix plot of random forest')
plt.colorbar(ax)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

print('\n Classification report \n',classification_report(y_test ,y_pred ))

```

## Second try in modeling -OverSampling+ Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

import sklearn
from sklearn.preprocessing import RobustScaler


y=train.y
cols = list(train)
cols.remove('y')
x =train[cols]



# from sklearn.preprocessing import Imputer
# imp = Imputer(missing_values='NaN', strategy='mean', axis=0)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25)
sm=SMOTE(random_state=0,kind = "regular")
                                                                            
ov_trainx,ov_trainy=sm.fit_sample(x_train,y_train)

##x_test=RobustScaler(quantile_range=(25, 75)).fit_transform(x_test)
#x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
#x_train, x_test, y_train, y_test =cross_validate.train_test_split(x,y,test_size=0.3)
print(len(ov_trainy),sum(ov_trainy))
                                                                            
classifier = RandomForestClassifier(max_depth=100, random_state=1,class_weight="balanced")  

classifier.fit(ov_trainx,ov_trainy)

# Predicting the Test set results
y_pred = classifier.predict(x_test)  
c_m = confusion_matrix(y_test,y_pred )
print('Logistic Regression:\nconfusion matrix\n', c_m,'\n\n')
ax=plt.matshow(c_m,cmap=plt.cm.Blues)
print('Confusion matrix plot of random forest')
plt.colorbar(ax)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

print('\n Classification report \n',classification_report(y_test ,y_pred ))
```
